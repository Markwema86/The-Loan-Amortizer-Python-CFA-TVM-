```python
# Project 1: Loan Amortizer
# Purpose: Calculate monthly payment and total interest

def calculate_monthly_payment(principal, annual_rate_percent, years):
  """
  Calculate the fixed monthly payment for a loan.

  Parameters:
  principal (float): The amount borrowed(e.g., 200000 for $200k)
  annual_rate_percent (float): Interest rate as a %(eg., 5.0 for 5%)
  years (int): Loan duration/term in years

  Returns:
  float: The fixed monthly payment amount.
  """
  # 1. Convert inputs to proper decimals and monthly frequency
  monthly_rate = (annual_rate_percent / 100) / 12
  total_payments = years * 12

  # 2. Apply the TVM formula
  # Edge case: If interest rate is )5, just divide principal by months
  if monthly_rate == 0:
    return principal / total_payments

  # PV * [ r(1+r)^n / ((1+r)^n - 1) ]
  numerator = monthly_rate * (1 + monthly_rate) ** total_payments
  denominator = ((1 + monthly_rate) ** total_payments) - 1
  monthly_payment = principal * (numerator / denominator)

  return monthly_payment

# --- Test the funtion ---
loan_amount = 200000 # $200,000 mortgage
rate = 5.0           # 5% interest
term = 30            # 30 years

payment = calculate_monthly_payment(loan_amount, rate, term)

print(f"Loan Amount: ${loan_amount:,.2f}")
print(f"Monthly Payment: ${payment:,.2f}")
```

```python
# Extension: How much did that house REALLY cost?

monthly_payment = calculate_monthly_payment(200000, 5.0, 30)
total_paid = monthly_payment * 30 * 12
total_interest = total_paid - 200000

print(f"Total Paid over 30 years: ${total_paid:,.2f}")
print(f"Total Interest Paid: ${total_interest:,.2f}")
print(f"Interest is {total_interest / loan_amount * 100:.1f}% of the original loan!")
```
