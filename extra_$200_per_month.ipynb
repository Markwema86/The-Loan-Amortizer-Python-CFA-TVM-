```python
# Project 1.1: Loan Amortizer
# Purpose:figure out how many months it takes to pay off the loan with an extra $200 per month
def calculate_months_with_extra_payment(principal, annual_rate_percent, years, extra_payment_per_month):
    """
    Calculates the number of months to pay off a loan with an extra monthly payment.

    Parameters:
    principal (float): The amount borrowed.
    annual_rate_percent (float): Interest rate as a %.
    years (int): Original loan term in years.
    extra_payment_per_month (float): The additional amount paid each month.

    Returns:
    tuple: (total_months, total_interest_paid)
    """
    monthly_rate = (annual_rate_percent / 100) / 12

    # Calculate the original required monthly payment
    original_monthly_payment = calculate_monthly_payment(principal, annual_rate_percent, years)

    # Total monthly payment including the extra amount
    actual_monthly_payment = original_monthly_payment + extra_payment_per_month

    current_principal = principal
    total_months = 0
    total_interest_paid = 0

    if actual_monthly_payment <= (principal * monthly_rate) and monthly_rate > 0: # Check if payment covers at least interest
        print("Warning: Monthly payment does not cover the interest. Loan will never be paid off.")
        return float('inf'), float('inf') # Loan will never be paid off

    while current_principal > 0:
        total_months += 1

        # Calculate interest for the current month
        interest_for_month = current_principal * monthly_rate
        total_interest_paid += interest_for_month

        # Subtract payment from principal (after interest is applied)
        payment_towards_principal = actual_monthly_payment - interest_for_month

        # Ensure we don't overpay the last bit of principal
        if current_principal - payment_towards_principal < 0:
            current_principal = 0
        else:
            current_principal -= payment_towards_principal

        # Just in case of floating point inaccuracies, if principal is very small, consider it paid
        if abs(current_principal) < 0.01:
            current_principal = 0

        # Safety break to prevent infinite loop for tiny payments not covering principal
        if total_months > (years * 12 * 2 + 100): # Allow for double the original term + buffer
            print("Loan taking an exceptionally long time to pay off. Check payment amount.")
            break

    return total_months, total_interest_paid

# --- Test the new function ---
loan_amount_extra = 200000
rate_extra = 5.0
term_extra = 30
extra_payment = 200

months_to_payoff, total_interest_paid_extra = calculate_months_with_extra_payment(
    loan_amount_extra, rate_extra, term_extra, extra_payment
)

original_monthly_payment = calculate_monthly_payment(loan_amount_extra, rate_extra, term_extra)
actual_payment_made = original_monthly_payment + extra_payment

print(f"\nLoan Amount: ${loan_amount_extra:,.2f}")
print(f"Original Monthly Payment: ${original_monthly_payment:,.2f}")
print(f"Extra Payment per Month: ${extra_payment:,.2f}")
print(f"Total Monthly Payment: ${actual_payment_made:,.2f}")
print(f"Months to Pay Off: {months_to_payoff}")
print(f"Total Interest Paid with Extra Payments: ${total_interest_paid_extra:,.2f}")

# Compare with original
original_total_paid = original_monthly_payment * term_extra * 12
original_total_interest = original_total_paid - loan_amount_extra

print(f"\nOriginal Total Interest Paid (30 years): ${original_total_interest:,.2f}")
print(f"Interest Saved by Extra Payments: ${original_total_interest - total_interest_paid_extra:,.2f}")
```

```python
# Calculate the real cost with extra payments
total_paid_with_extra = actual_payment_made * months_to_payoff
total_interest_with_extra = total_interest_paid_extra
interest_percentage_with_extra = (total_interest_with_extra / loan_amount_extra) * 100

print(f"\n--- Real Cost with Extra Payments ---")
print(f"Total Paid (with extra payments): ${total_paid_with_extra:,.2f}")
print(f"Total Interest Paid (with extra payments): ${total_interest_with_extra:,.2f}")
print(f"Interest is {interest_percentage_with_extra:.1f}% of the original loan!")
```
