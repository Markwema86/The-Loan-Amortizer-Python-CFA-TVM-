```python
# Project 2: Savings Goal Calculator (Future Value of a Lump Sum)
# Purpose: Calculate the future value of a one-time investment.

def calculate_future_value_lump_sum(principal, annual_rate_percent, years):
    """
    Calculates the future value of a lump sum investment.

    Parameters:
    principal (float): The initial amount invested (present value).
    annual_rate_percent (float): The annual interest rate as a %.
    years (int): The number of years the money is invested.

    Returns:
    float: The future value of the investment.
    """
    # Convert annual rate to decimal and years to periods (assuming annual compounding for simplicity)
    rate_decimal = annual_rate_percent / 100
    num_periods = years

    # Future Value (FV) = PV * (1 + r)^n
    future_value = principal * ((1 + rate_decimal) ** num_periods)

    return future_value

# --- Test the new function ---
savings_principal = 1000000  # Initial lump sum investment
savings_rate = 7.0       # 7% annual interest rate
savings_years = 10       # Invested for 10 years

future_value_result = calculate_future_value_lump_sum(savings_principal, savings_rate, savings_years)

print(f"\n--- Savings Goal Calculator ---")
print(f"Initial Investment: ${savings_principal:,.2f}")
print(f"Annual Interest Rate: {savings_rate:.1f}%")
print(f"Investment Period: {savings_years} years")
print(f"Future Value of Investment: ${future_value_result:,.2f}")
```
